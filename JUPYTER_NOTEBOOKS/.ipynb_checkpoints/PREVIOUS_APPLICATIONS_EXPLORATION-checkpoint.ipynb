```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings('ignore')
print("Directorio actual:", os.getcwd())
# Have all columns appear when dataframes are displayed.
pd.set_option('display.max_columns', None) 
# Have 100 rows appear when a dataframe is displayed
pd.set_option('display.max_rows', 500)
# Display dimensions whenever a dataframe is printed out.
pd.set_option('display.show_dimensions', True)

previous_app= pd.read_csv(r'/home/yeray/home-credit-default-risk/previous_application.csv')
```

```python
#Al igual que en Bureau tenemos una fila por cada aplicación de crédito anterior de los clientes de Home Credit.
#Vamos a ver la diferencia en el número de filas
print(previous_app['SK_ID_CURR'].nunique())
print(previous_app['SK_ID_PREV'].nunique())
```

```python
#Veamos cuales son las varibales categóricas 
variables_categoricas = previous_app.select_dtypes('object', 'category').columns
print("Hay ", len(variables_categoricas), "variables categoricas y son: " ,variables_categoricas)


#Vamos a ver los valores de cada una de estas para prevenir que haya valores que no tengan sentido
for var in variables_categoricas:
    print(f"Valores únicos en la columna {var}: ")
    print(previous_app[var].unique())
#Además de XNA que ya sabiamos que había que reemplazar, XAP también habrá de ser reemplazado por NaN
```

```python
#Vamos a ver las variables numéricas
variables_numericas = previous_app.select_dtypes('float64', 'int64').columns
print("Hay ", len(variables_numericas), "variables numéricas y son: " ,variables_numericas)

#Veamos su distribución de manera gráfica
for var in variables_numericas:
    sns.histplot(previous_app[var])
    plt.title(f"Distribución de la variable {var}")
    plt.show()
```
