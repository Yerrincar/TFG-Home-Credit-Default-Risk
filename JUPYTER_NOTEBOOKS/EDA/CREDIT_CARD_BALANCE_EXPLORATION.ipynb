```python
#Vamos a ver la relación entre el número de clientes y los gastos de la tarjeta de crédito
print(credit_card_balance['SK_ID_PREV'].nunique())
print(credit_card_balance['SK_ID_CURR'].nunique())

#La información de la que disponemos de cada fila es:
#-MONTHS_BALANCE: Mes relativo a la fecha de la aplicación; -1 indica el mes más reciente.
#-AMT_BALANCE: Saldo de la tarjeta en ese mes (deuda pendiente).
#-AMT_CREDIT_LIMIT_ACTUAL: Límite de crédito disponible ese mes.
#-AMT_DRAWINGS_ATM_CURRENT: Monto de retiros en cajero automático (ATM) durante el mes.
#-AMT_DRAWINGS_CURRENT: Monto total de extracciones (ATM, POS, etc.) en el mes.
#-AMT_DRAWINGS_OTHER_CURRENT: Monto de otras extracciones no clasificadas en ATM o POS.
#-AMT_DRAWINGS_POS_CURRENT: Monto de compras en POS (bienes/servicios) durante el mes.
#-AMT_INST_MIN_REGULARITY: Pago mínimo exigido de la tarjeta en ese mes.
#-AMT_PAYMENT_CURRENT: Cuánto pagó el cliente realmente en ese mes.
#-AMT_PAYMENT_TOTAL_CURRENT: Total pagado (parcial, intereses, etc.) en ese mes.
#-AMT_RECEIVABLE_PRINCIPAL: Principal pendiente de cobro en la tarjeta.
#-AMT_RECIVABLE: Monto a cobrar (puede incluir intereses/comisiones) ese mes.
#-AMT_TOTAL_RECEIVABLE: Total a cobrar (principal + intereses + comisiones).
#-CNT_DRAWINGS_ATM_CURRENT: Número de extracciones ATM en el mes.
#-CNT_DRAWINGS_CURRENT: Número total de extracciones (ATM, POS...) en el mes.
#-CNT_DRAWINGS_OTHER_CURRENT: Número de extracciones “other” en el mes.
#-CNT_DRAWINGS_POS_CURRENT: Número de compras POS en el mes.
#-CNT_INSTALMENT_MATURE_CUM: Cantidad de cuotas de la tarjeta ya pagadas.
#-NAME_CONTRACT_STATUS: Estado del crédito/tarjeta (Active, Closed, etc.).
#-SK_DPD: Días de atraso (Days Past Due) ese mes.
#-SK_DPD_DEF: Días de atraso con tolerancia (pequeñas deudas ignoradas).
```

```python
#Vamos a ver la distribución de las variables numéricas
variables_numericas = credit_card_balance.select_dtypes(include=[np.number]).columns
variables_numericas = variables_numericas.drop(['SK_ID_CURR', 'SK_ID_PREV'])

for var in variables_numericas:
    plt.figure(figsize=(10,6))
    sns.displot(credit_card_balance[var], bins=30, kde=True, color='blue')
    plt.title(f'Distribución de la variable {var}')
    plt.show()
```
