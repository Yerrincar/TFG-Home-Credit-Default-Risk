```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings('ignore')
print("Directorio actual:", os.getcwd())
# Have all columns appear when dataframes are displayed.
pd.set_option('display.max_columns', None) 
# Have 100 rows appear when a dataframe is displayed
pd.set_option('display.max_rows', 500)
# Display dimensions whenever a dataframe is printed out.
pd.set_option('display.show_dimensions', True)

pos_cash = pd.read_csv(r'/home/yeray/home-credit-default-risk/POS_CASH_balance.csv')
```

```python
#Información general del dataset
print(pos_cash.info())	
print(pos_cash.dtypes)
print(pos_cash.describe())
print(pos_cash.head(5))
```

```python
#Vamos a ver los valores de esa columna categórica
print(pos_cash['NAME_CONTRACT_STATUS'].value_counts())
```

```python
#Variables numéricas
variables_numericas = pos_cash.select_dtypes(include = ['float64', 'int64']).columns
print("Hay ", len(variables_numericas), "variables numéricas y son: " ,variables_numericas)

#Veamos su distribución de manera gráfica
for var in variables_numericas:
    sns.histplot(pos_cash[var])
    plt.title(f"Distribución de la variable {var}")
    plt.show()
```
